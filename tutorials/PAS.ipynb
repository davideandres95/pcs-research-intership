```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#chi = [-3.0, -1.0, 1.0, 3.0]
#chi = [-1.5, -1., -0.5, 0.5, 1.0, 1.5]  # alphabet
chi = np.arange(-0.9, 1.1, 0.45)
shapes = np.arange(0, 10, 1)
print(chi)
print(shapes)
```

```python
def px_mb(shaping, alphabet):
    denominator = sum([ np.exp(-shaping*i*i) for i in alphabet])
    numerator = np.exp(-shaping*np.power(alphabet,2))
    return numerator/denominator
    
```

```python
def dif_entropy(px):
    return -np.mean(np.log2(px))
```

```python
x = chi
for shape in shapes:
    y = px_mb(shape,chi)
    plt.plot(x, y, label="shape="+str('{0:.3g}'.format(shape)))

plt.grid()
plt.legend()
plt.xlabel('x')
plt.ylabel('$P_{X} (x)$')
plt.show()
```

```python
x = chi
y = px_mb(0.05, chi)
plt.plot(x, y, 'ro')
plt.show()

print(sum(y))
```

```python
fig, ax1 = plt.subplots()

h = [dif_entropy(px_mb(shape, chi)) for shape in shapes]
var = [np.var(px_mb(shape, chi)) for shape in shapes]

ax2 = ax1.twinx()
ax1.plot(shapes, h, 'g-')
ax2.plot(shapes, var, 'b-')

ax1.set_xlabel('Shape')
ax1.set_ylabel('H(X)', color='g')
ax2.set_ylabel('Var[X]', color='b')

plt.show()
```
